# 05 – Results Analysis

Load saved simulation results and produce publication-quality visualisations.

```python
import sys, pathlib
sys.path.insert(0, str(pathlib.Path.cwd().parent))
```

```python
from hotelling.utils.data_io import load_json
from hotelling.simulation.metrics import aggregate_results
import pathlib

results_path = pathlib.Path('../results/general/general_results.json')

if results_path.exists():
    results = load_json(results_path)
    agg = aggregate_results(results)
    print(f'Loaded {len(results)} steps. Firms: {list(agg["firms"].keys())}')
else:
    print(f'Results file not found at {results_path}. Run experiments/general_hotelling.py first.')
    agg = None
```

```python
if agg:
    try:
        import matplotlib.pyplot as plt
        from hotelling.visualization.plots import (
            plot_market_shares, plot_price_evolution,
            plot_firm_trajectories
        )

        fig, axes = plt.subplots(1, 3, figsize=(18, 5))
        plot_market_shares(agg, ax=axes[0])
        plot_price_evolution(agg, ax=axes[1])
        plot_firm_trajectories(agg, ax=axes[2])
        plt.suptitle('Simulation Results Analysis', fontsize=14)
        plt.tight_layout()
        plt.show()
    except ImportError:
        print('Install matplotlib: pip install matplotlib')
```

```python
if agg:
    try:
        from hotelling.visualization.interactive import (
            interactive_market_shares, interactive_price_evolution
        )
        fig = interactive_market_shares(agg)
        fig.show()
    except ImportError:
        print('Install plotly for interactive charts: pip install plotly')
```
