# 02 – General Hotelling Simulation

This notebook runs a full simulation with two naive (centre-seeking) firms
and analyses the results.

```python
import sys, pathlib
sys.path.insert(0, str(pathlib.Path.cwd().parent))
```

```python
from hotelling.agents.naive_agent import NaiveAgent
from hotelling.simulation.config import SimulationConfig
from hotelling.simulation.engine import SimulationEngine
from hotelling.simulation.metrics import aggregate_results

cfg = SimulationConfig(n_steps=150, n_consumers=2000, seed=42, log_every=5)
engine = SimulationEngine(config=cfg)

firms = [
    NaiveAgent('Firm_A', location=(0.1, 0.5), strategy='center', seed=0),
    NaiveAgent('Firm_B', location=(0.9, 0.5), strategy='center', seed=1),
]

engine.setup(firms)
results = engine.run()
agg = aggregate_results(results)
print('Simulation complete – recorded steps:', len(results))
```

```python
print(engine.market.summary())
```

```python
try:
    import matplotlib.pyplot as plt
    from hotelling.visualization.plots import (
        plot_market_shares, plot_price_evolution, plot_firm_trajectories, plot_snapshot
    )

    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    plot_market_shares(agg, ax=axes[0])
    plot_price_evolution(agg, ax=axes[1])
    plot_firm_trajectories(agg, ax=axes[2])
    plt.suptitle('General Hotelling Simulation Results', fontsize=14)
    plt.tight_layout()
    plt.show()

    plot_snapshot(engine.market, title='Final Market State')
    plt.show()
except ImportError:
    print('Install matplotlib: pip install matplotlib')
```
