# 01 – Getting Started with the Hotelling Toolkit

This notebook provides a guided tour of the `hotelling` package.
By the end you will have run a simple two-firm simulation and visualised the results.

```python
# If running from the repo root without installing, add the package to the path
import sys, pathlib
sys.path.insert(0, str(pathlib.Path.cwd().parent))
```

```python
import hotelling
print('hotelling version:', hotelling.__version__)
```

## 1. Create a City

```python
from hotelling.core.city import City

city = City(width=1.0, height=1.0, n_consumers=1000, seed=42)
print(city)
print('Consumer array shape:', city.consumer_locations.shape)
```

## 2. Create Firms

```python
from hotelling.core.firm import Firm

firm_a = Firm(firm_id='A', location=(0.25, 0.5), price=1.0)
firm_b = Firm(firm_id='B', location=(0.75, 0.5), price=1.0)
print(firm_a)
print(firm_b)
```

## 3. Build a Market and Resolve

```python
from hotelling.core.market import Market

market = Market(city=city, firms=[firm_a, firm_b], transport_cost=1.0)
shares = market.resolve()
print(market.summary())
```

## 4. Visualise the Snapshot (requires matplotlib)

```python
try:
    import matplotlib.pyplot as plt
    from hotelling.visualization.plots import plot_snapshot
    plot_snapshot(market, title='Initial Market Snapshot')
    plt.show()
except ImportError:
    print('Install matplotlib to see the plot: pip install matplotlib')
```
